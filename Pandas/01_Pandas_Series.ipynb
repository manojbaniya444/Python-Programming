## Pandas Series
- A column in a table is a `Series`
- 1-Dimensional array holding data of any type

```python
import pandas as pd
```

```python
# Series from list
country = ["Nepal", "India", "China", "Japan"]
pd.Series(country)
```

```python
# series from list another 
age = [10, 20, 30, 40, 50, 60, 70, 80]
pd.Series(age)
```

```python
# giving custom index
marks = [98, 96, 30, 40]
subjects = ["Math", "Science", "Nepali", "English"]
pd.Series(marks, index=subjects)
```

```python
# giving name to series
pd.Series(marks, index=subjects, name="Marks Series")
```

```python
# Series from dictionary
marks = {
    "Math": 98,
    "Science": 96,
    "Nepali": 30,
    "English": 40
}

marks_data = pd.Series(marks)
```

```python
# plotting the marks series
marks_data
```

```python
marks_data.plot() # plotting the data
```

```python
marks_data.value_counts()
```

```python
marks_data.value_counts().head(2).plot(kind="bar")
```

```python
marks_data.value_counts().head(2).plot(kind="pie")
```

## Some Attributes of Series of Pd

```python
size = marks_data.size # number of elements in series
name = marks_data.name # name of series
is_unique = marks_data.is_unique # check if the series has unique values
index = marks_data.index # index value return garxa
values = marks_data.values # returns all values

print(f"Size: {size}")
print(f"Name: {name}")
print(f"Is Unique: {is_unique}")
print(f"Index: {index}")
print(f"Values: {values}")
```

## Series Methods

```python
series = pd.Series(["Messi", "Neymar", "Ronaldo", "Zlatan", "Lewandowski", "Mbappe", "Haaland", "Bellingham"], name="Football Players")
```

```python
# top 5
series.head()
```

```python
type(series)
```

```python
# bottom 5
series.tail()
```

```python
# get random
series.sample()
```

```python
# get random n
series.sample(5)
```

```python
# count the total unique values
series.value_counts()
```

```python
# sort: ascending
series.sort_values()
```

```python
# sort: descending
series.sort_values(ascending=False)
```

```python
# change to original
series.sort_values(ascending=False, inplace=True)
```

```python
series
```

```python
# sorting index
series.sort_index()
```

```python
# count all values which are non null
series.count()
```

```python
# sum all values
marks_data.sum()
```

```python
# product of all values in a series
marks_data.product()
```

```python
# mean median std and var
mean = marks_data.mean()
median = marks_data.median()
std = marks_data.std()
variance = marks_data.var()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std}")
print(f"Variance: {variance}")
```
