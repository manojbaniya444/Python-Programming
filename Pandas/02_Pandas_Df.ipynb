## Filtering the data

```python
ipl_data = pd.read_csv("./Data/ipl-matches.csv")
```

```python
ipl_data.columns
```

```python
ipl_data.sample(5)
```

```python
mask = ipl_data["MatchNumber"] == "Final"
```

```python
df = ipl_data[mask]
df.iloc[:5]
```

```python
df[["Season", "Team1", "Team2"]]
```

```python
# How many superover
ipl_data[ipl_data["SuperOver"] == "Y"].shape[0]
```

```python
# How many times RCB won on Delhi ground
mask_city = ipl_data["City"] == "Delhi"
mask_team = ipl_data["WinningTeam"] == "Royal Challengers Bangalore"

ipl_data[mask_city & mask_team].shape[0]
```

```python
# If toss winning team also won the match
mask = ipl_data["TossWinner"] == ipl_data["WinningTeam"]

toss_won_match_won = ipl_data[mask].shape[0]
total_matches = ipl_data.shape[0]

percent_won_by_toss_won = (toss_won_match_won / total_matches) * 100

print(f"Percent when the toss winning team also win the matches: {percent_won_by_toss_won: .2f}")
```
