## Pandas Group By Method

```python
import pandas as pd
import numpy as np
```

```python
sample_data = {
    "col1": [1, 2, 2, 1, 5, 6],
    "col2": ["A", "B", "B", "A", "A", "C"]
}

sample_df = pd.DataFrame(sample_data)
```

```python
sample_df
```

```python
sample_df.groupby("col1").sum()
```

```python
sample_df.groupby("col1").min()
```

```python
sample_df.groupby("col2").sum()["col1"]
```

```python
movies_df = pd.read_csv("./DATA/imdb-top-1000.csv")
```

```python
movies_df.head()
```

```python
genre_group = movies_df.groupby("Genre")
```

```python
genre_group
```

```python
genre_group.sum()
```

```python
genre_group.min().head(2)
```

```python
movies_df.columns
```

```python
# Top 3 movies genre by total earning
movies_df.groupby("Genre").sum()["Gross"].sort_values(ascending=False).head(3)
```

```python
# Genre with highest average rating
movies_df.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
movies_df["IMDB_Rating"].mean()
```

```python
# director with most popularity
movies_df.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(1)
```

```python
# highest rated movie of each genre
highest_rated_movies = genre_group.apply(lambda x: x.loc[x['IMDB_Rating'].idxmax()])
highest_rated_movies
```

```python
# No of moveies done by each actor
len(movies_df.groupby("Genre"))
```

```python
# Aggregation
genre_group.agg(
    {
        "Runtime": "mean",
        "IMDB_Rating": "mean",
        "Gross": "sum",
        "No_of_Votes": "mean",
        "Metascore": "min"
    }
)
```

```python
movies_df.groupby("Genre").agg(
    {
        "Runtime": ["mean", "max"],
        "Gross": ["sum", "max"],
        "Metascore": "max",
        "No_of_Votes": "mean"
    }
)
```

## Looping

```python
for group, data in genre_group:
    print(group, data)
    break
```

```python
genre_group.apply(min, include_groups=False)
```

```python
# Actor name starts with A
def actorNameWithA(group):
    """Actor name starting with A"""
    row = movies_df["Star1"].str.startswith("A").sum()
    return row

genre_group.apply(actorNameWithA)
```

```python
# rank each group in the group
def rankMovie(group):
    a = group["IMDB_Rating"].rank(ascending=False)
    movies_df["rank_genre"] = a
    return group
```

```python
movies_df.groupby("Genre").apply(rankMovie).head(10)
```

```python
duo = movies_df.groupby(["Director", "Star1"])
```

```python
duo.size()
```
