```python
def func(*args, **kwargs):
  print(args)
  print(kwargs)
  print(next(iter(args)))
  print(kwargs.get("a", 0))
```

```python
func(1,2,3,**{"a": 1, "b": 10})
```

```python
def verify_auth(func):
  def wrapper(*args, **kwargs):
    print("Before the function")
    func(*args, **kwargs)
    print("After the function")
  return wrapper
```

```python
@verify_auth
def test(user):
  print(f"Authenticating the user: {user}")
```

```python
test("Test")
```

### OOP

```python
from abc import ABC, abstractmethod
```

```python
class BaseUser(ABC):
  @abstractmethod
  def register():
    pass
```

```python
class Employee(BaseUser):
  def __init__(self, username):
    self.username = username
  
  def register(self):
    print("register")
```

```python
user1 = Employee("Test")
```

```python
user1.register()
```

```python
class SimpleClass():
  def __init__(self):
    print("Class init")
    
  @staticmethod
  def say_hi():
    print("Hi")
    
  def test(self):
    SimpleClass.say_hi()
```

```python
test1 = SimpleClass()
```

```python
test1.say_hi()
```

```python
test1.test()
```

```python
SimpleClass.say_hi()
```

### Inheritance

```python
class Base():
  def __init__(self):
    print("Base class init")
    
  def base_method(self):
    print("Hi from the base class method test")
```

```python
class Base2():
  def __init__(self):
    print("Hi from the base class 2")
    
  def base_method_2(self):
    print("Hi from the base method 2")
```

```python
class Derived(Base):
  def __init__(self):
    print("Derived class init test")
    
  def derived_method(self):
    print("Hi from the derived method test")
```

```python
derived_test = Derived()
```

```python
derived_test.base_method()
```

```python
class DerivedMultiple(Base, Base2):
  def __init__(self):
    print("derived the multiple base class")
    
  def derived_method_2(self):
    print("Hi from the derived method 2 test")
```

```python
derived_test_other = DerivedMultiple()
```

```python
derived_test_other.base_method()
derived_test_other.base_method_2()
```

```python
class OtherDerived(DerivedMultiple):
  pass
```

```python
derived_test_other_2 = OtherDerived()
```

```python
derived_test_other_2.base_method()
```
