## Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2D Line Plot

```python
price = [48000, 10000, 20000, 30000, 60000]
year = [2012, 2013, 2018, 2019, 2020]

plt.plot(year, price, linestyle="dashdot", linewidth=1) # dash solid
plt.scatter(year, price)
plt.xlabel("Year")
plt.ylabel("Price")
```

```python
runs = pd.read_csv("./data/sharma-kohli.csv")
```

```python
runs
```

```python
plt.plot(runs["index"], runs["V Kohli"], label="Kohli", linestyle="dashed")
plt.plot(runs["index"], runs["RG Sharma"], label="Rohit", linestyle="dashdot")
plt.title("Kohli vs Rohit")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
```

```python

x = np.arange(10)
y = np.power(x, 2)

plt.style.use("default")

plt.plot(x, y, linestyle="dashed")
plt.xlabel("X")
plt.ylabel("X Squared")
plt.title("X vs X squared")
plt.grid() # activate grid
plt.scatter(x, y, c='orange')
plt.savefig("./figures/plot_fig.png")
```

## Scatter Plot

```python
x = np.linspace(-10, 10, 50)
y = 10 * x + 3 + np.random.randint(0, 300, 50)

plt.scatter(x, y)
```

## Pie Chart

```python
data = [80, 60, 100]
marks= ["subject a", "subject b", "subject c"]
plt.pie(data, labels=marks, colors=["green", "orange", "cyan"], autopct="%0.1f%%", explode=[0.1, 0.08, 0.2], shadow=True)
plt.savefig("./figures/pie_plot.png")
```
